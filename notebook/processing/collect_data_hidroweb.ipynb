- Salvando o dataset completo em um unico documento csv

```python
import csv
import os

def save_data(datasets):
    csv_file = "/Users/arthurhenrique/Documentos/pessoal/faculdade/documentos/tcc/repository/machine-learning-water-quality/notebook/output/output_amazonia.csv"

    # Verifica se o diretório existe e cria se necessário
    os.makedirs(os.path.dirname(csv_file), exist_ok=True)

    keys = datasets[0].keys()
    file_exists = os.path.isfile(csv_file)

    with open(csv_file, 'a', newline='') as file:
        dict_writer = csv.DictWriter(file, fieldnames=keys)
        if not file_exists:
            dict_writer.writeheader()
        
        dict_writer.writerows(datasets)
```
