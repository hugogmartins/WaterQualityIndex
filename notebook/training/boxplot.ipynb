```python

def graficoHistograma(data, xlabel, title):
    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
    plt.title(f'Histograma da {title}')
    plt.xlabel(xlabel)
    plt.ylabel('Frequência')
    plt.grid(True)
    plt.show()

for colunm in data.columns:
    graficoHistograma(data[colunm], colunm, colunm)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def graficoDensidade(data, xlabel, title):
    plt.figure(figsize=(10, 6))
    sns.kdeplot(data, shade=True, color='blue')
    plt.title(f'Gráfico de Densidade da {title}')
    plt.xlabel(xlabel)
    plt.ylabel('Densidade')
    plt.grid(True)
    plt.show()

for colunm in data.columns:
    graficoDensidade(data[colunm], colunm, colunm)
```
